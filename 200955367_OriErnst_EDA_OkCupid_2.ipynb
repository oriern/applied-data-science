```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('profiles.csv')
```

Here we can see that in taking drugs, men age is higher than wemen, and 'often' age is lower than 'sometimes' age. 

```python
df_drugs_notNan= df[pd.notnull((df['drugs']))]
df_drugs = df_drugs_notNan[df_drugs_notNan['drugs']!='never']
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x='age', y='drugs',hue='sex',data=df_drugs)
```

```python
plt.figure(figsize=(15, 6))
sns.factorplot(x='drugs',y='age',col='sex',col_wrap=4,data=df_drugs,kind='bar')
```

A graph showing height distribution for each sex 

```python
df_height_notNan= df[pd.notnull((df['height']))]
plt.figure(figsize=(15, 6))
sns.distplot(df_height_notNan[df_height_notNan['sex']=='m']['height'])
sns.distplot(df_height_notNan[df_height_notNan['sex']=='f']['height'])
```

statistics of diet for each ethnics group 

```python
sns.factorplot(x='diet',col='ethnicity',col_wrap=4,data=df,kind='count')
```

heatmap for correlation between smokes and like pets.
We can see, that there is no correlation between the two. Each row is quite uniform.

```python
SmokePets =  df.groupby(by='smokes')['pets'].value_counts(normalize=True).unstack()
plt.figure(figsize=(12,6))
sns.heatmap(SmokePets,cmap='viridis')
```
