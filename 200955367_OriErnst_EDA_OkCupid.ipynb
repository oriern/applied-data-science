```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
df=pd.read_csv('profiles.csv')
```

Here we can see the age density

```python
sns.kdeplot(df['age'])
```

There are more men than weman in the website, but in jews it is almost equal.

```python
df['religion1']=df['religion'].apply(lambda x: str(x))
df['religion1']=df['religion1'].apply(lambda x: x.split()[0])
sns.countplot(x='religion1',hue='sex',data=df)

```

Height mean is under 70, and there are many outliers.

```python
sns.boxplot(df['height'])
```

age vs height

```python
byAge=df.groupby('height')
c=pd.DataFrame(byAge['age'].mean())
#sns.lmplot(x='x',y='age',data=c)
c.reset_index( inplace=True)
#c
sns.jointplot(x='height',y='age',data=c)
```
